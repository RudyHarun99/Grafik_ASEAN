```python
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
```

```python
db=mysql.connector.connect(
    host='localhost',
    port=3306,
    user='root',
    passwd='sekutu99',
    database='world'
)
c=db.cursor()
```

```python
query="select name, population from country where region='Southeast Asia' order by name"
dfPopulasi=pd.read_sql(query,db)
dfPopulasi
```

```python
plt.figure('Populasi',figsize=(10,5))
plt.bar(dfPopulasi.name,dfPopulasi.population,color=['cyan','green','blue','red','magenta','black','yellow','purple','orange','gold','grey'])
plt.xticks(rotation=45)
plt.xlabel('Negara')
plt.ylabel('Populasi')
plt.title('Populasi Negara ASEAN')
for i in dfPopulasi.population:
    plt.text(dfPopulasi[dfPopulasi.population==i].name,i,i,ha='center',va='bottom')
plt.show()
```

```python
x,y,z=plt.pie(dfPopulasi.population,labels=dfPopulasi.name,textprops={'size':10,'color':'r'},autopct='%1.1f%%',radius=2)
for i in z:
    i.set_color('w')
plt.show()
```

```python
query="select name, gnp from country where region='Southeast Asia' order by name"
dfGNP=pd.read_sql(query,db)
dfGNP
```

```python
plt.figure('GNP',figsize=(10,5))
plt.bar(dfGNP.name,dfGNP.gnp,color=['cyan','green','blue','red','magenta','black','yellow','purple','orange','gold','grey'])
plt.xticks(rotation=45)
plt.xlabel('Negara')
plt.ylabel('GNP')
plt.title('Pendapatan Bruto Nasional ASEAN')
for i in dfGNP.gnp:
    plt.text(dfGNP[dfGNP.gnp==i].name,i,i, ha = 'center', va = 'bottom')
plt.show()
```
